```python
import fitz  # PyMuPDF
import ocrmypdf
import re

# Step 1. OCR the scanned PDF (skip if already searchable)
ocrmypdf.ocr("CMSP-Licensing-and-Supervision-Directive-No-980-2024.pdf", "output_searchable.pdf", language="eng")

# Step 2. Extract text with PyMuPDF
doc = fitz.open("output_searchable.pdf")
pages_text = []

for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text("text")

    # Step 3. Basic cleanup
    text = re.sub(r'-\n', '', text)              # join hyphenated line breaks
    text = re.sub(r'\n+', '\n', text)            # collapse multiple newlines
    text = re.sub(r'(?<!\n)\n(?!\n)', ' ', text) # join broken lines within a paragraph
    text = re.sub(r'\s+', ' ', text)             # normalize spaces
    text = text.strip()

    pages_text.append(text)

# Step 4. Save cleaned text as a .txt file
full_text = "\n\n".join(pages_text)  # separate pages cleanly

with open("standardized.txt", "w", encoding="utf-8") as f:
    f.write(full_text)

print("Saved cleaned OCR output to standardized.txt")
```
